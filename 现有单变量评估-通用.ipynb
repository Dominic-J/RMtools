# 分箱

:param df: dataframe 要求dtype = float   
:param target: Y标签  
:param max_bin: 最大分箱箱数  
:param method: mono模型分箱方法; tree决策树分箱; widthequal 等距; depthequal 等频率  
:param alg_method: 计算指标 'iv','gini','entropy'信息熵  

```python
import woe_bin
from eval_thirdscore import auto_binning
from excel_format_lift import reset_dataframe, excel_format_lift
```

```python
#bin_path = path+'分箱结果/10月20后/'
bin_path = path
```

```python
df_bin = df_sample
```

```python
# 是否到期统计标签
y_name = ['is_init_credit']
# 是否逾期统计标签
y_var = ['is_cre']   
```

```python
# 是否到期统计标签
y_name = ['agr_fpd10_cnt','agr_fpd30_cnt','agr_mob04_30_cnt']
# 是否逾期统计标签
y_var = ['def_fpd10_cnt','def_fpd30_cnt','def_mob04_30_cnt']     
```

```python
for i in range(len(y_var)):
    i_name = y_name[i]
    i_var = y_var[i]
    bin_col = score + [i_var]
    #woe = auto_binning(df_bin[df_bin[i_name] >= 1][bin_col], i_var, 10, 'depthequal', 'iv')[1]   # 参数分别是数据源，y标签，最大分箱箱数，分箱方法，计算指标
    woe = auto_binning(df_bin[df_bin[i_name] >= 1][bin_col], i_var, 10, 'auto', 'iv')[1]
    
    # 按照变量名排序,按分箱升序
    woe = woe.sort_values(by=['varname', 'bin'], ascending=[True, True]).reset_index(drop=True)
    result = reset_dataframe(woe)
    final_name = "10月20日后_最优10分箱_" + i_var + ".xlsx"
    writer = pd.ExcelWriter(bin_path+final_name)
    result.to_excel(bin_path+final_name, index=False, sheet_name=i_var)
    excel_format_lift(bin_path, final_name, store_path=bin_path)
```
